```python
class a:

    def __init__(self):
        print("in A Init")

    def feature1(self):
        print("feature 1-A working")

    def feature2(self):
        print("feature 2 working")

class b:

    def __init__(self):
        super().__init__()
        print("in B Init")
    
    def feature3(self):
        print("feature 3-B working")

    def feature4(self):
        print("feature 4 working")

class c(a,b):
    def __init__(self):
        super().__init__()
        print("in C init")


a1 = c()
a1.feature1()
```

```python
class a:

    def __init__(self):
        print("in A Init")

    def feature1(self):
        print("feature 1-A working")

    def feature2(self):
        print("feature 2 working")

class b:

    def __init__(self):
        super().__init__()
        print("in B Init")
    
    def feature3(self):
        print("feature 3-B working")

    def feature4(self):
        print("feature 4 working")

class c(a,b):
    def __init__(self):
        super().__init__()
        print("in C init")


a1 = c()
a1.feature1()
a1.feature3()
```
